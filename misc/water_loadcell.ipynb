```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load cam 1 excels
path_fc = r'/home/merge/merge/water'

df_allsens = pd.read_csv(path_fc+'/parsed_data_7_24_15_10_30.csv', index_col=None, header=0)
df_allsens
```

```python
df_loads = df_allsens
arr_cells = ['C0', ' C2', ' C4']

plt.figure(figsize=(10,8))
for cell in arr_cells:
    plt.plot(df_loads['T_State'], df_loads[cell], label=cell)
    plt.legend()
```

```python
import matplotlib as mpl
ax = plt.gca()
# plt.rcdefaults()
plt.rcParams["font.family"] = "serif"
plt.rcParams['font.size'] = 10
ax.xaxis.set_tick_params(which='major', direction='in', bottom='on')
ax.yaxis.set_tick_params(which='major', direction='in', left='on')

plt.plot(df_loads["Time"], df_loads["Loadcell 1"], label="Perforated", color='black')
plt.plot(df_loads["Time"], df_loads["Loadcell 3"], label="Control", color='dimgrey')
plt.plot(df_loads["Time"], df_loads["Loadcell 5"], label="Ringed", color='grey')
plt.xlabel('Time (s)', labelpad=10)
plt.ylabel('Force', labelpad=10)
plt.legend(frameon=False)
plt.figure(dpi=200)
plt.savefig('fc72'+'loadcell'+'.png', dpi=300, transparent=False, bbox_inches='tight')
plt.show()
```
