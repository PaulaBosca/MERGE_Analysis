```python
h = 16.192    #chamber height

df_list = [df_c1p1, df_c1p2, df_c1p3, df_c1p4, df_c1p5, df_c1p6, df_c1p7, df_c1p8, df_c1p9,
           df_c2p1, df_c2p2, df_c2p3, df_c2p4, df_c2p5, df_c2p6, df_c2p7, df_c2p8, df_c2p9, 
           df_c3p1, df_c3p2, df_c3p3, df_c3p4, df_c3p5, df_c3p6, df_c3p7, df_c3p8, df_c3p9]

def addingNab(df):
    mask = df.columns.str.contains('Slice\s[^4]\sMax')
    df_masked = df.loc[:,mask]
    df["Nab25"] = df_masked.gt(0.25*h).sum(axis=1)
    df["Nab35"] = df_masked.gt(0.35*h).sum(axis=1)
    df["Nab45"] = df_masked.gt(0.45*h).sum(axis=1)
    df["Nab50"] = df_masked.gt(0.50*h).sum(axis=1)
    df["Nab75"] = df_masked.gt(0.75*h).sum(axis=1)
    df["Nab95"] = df_masked.gt(0.95*h).sum(axis=1)
    return df

for df in df_list: 
    df = addingNab(df)
```

```python
def addingMaxHeight(df):
    df_masked = df.filter(regex='Slice*')
    df["MaxHeight"] = df_masked.max(axis=1)

for df in df_list:
    df = addingMaxHeight(df)
```

```python
#testing a fit for one of the plots 
# testparab = 7 #indexing starts at 0
# parab = testparab
# cam1 = df_cam1[parab]
# cam2 = df_cam2[parab]
# cam3 = df_cam3[parab]
# cams = [cam1, cam2, cam3]

polyorder = 15

import numpy.polynomial.polynomial as poly
colors = ['k', 'b','r']

for parab in range(0,9):
    cam1 = df_cam1[parab]
    cam2 = df_cam2[parab]
    cam3 = df_cam3[parab]
    cams = [cam1, cam2, cam3]

    fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(ncols=2, nrows=2, figsize=(14,12))
    fig.suptitle("Parab %i"%(parab+1))
    fig.subplots_adjust(top=0.95)
    df = cam1
    coefs = poly.polyfit(df["Frame"], df["MaxHeight"], polyorder)
    x_new = np.linspace(df["Frame"].iloc[1], df["Frame"].iloc[-1], num=len(df["Frame"])*10)
    ffit = poly.polyval(x_new, coefs)
    fderiv = poly.polyval(x_new, np.polyder(coefs))
    
    
    ax1.plot(x_new, ffit, label="Chamber 1 (Perforated Slat)", color=colors[0])
    ax4.plot(x_new, fderiv, label="Chamber 1 fit deriv", color=colors[0])
#     ax4.plot(x_new, ffit, label="Cam "+ str(df["cam"].iloc[0]), color=colors[0])
    ax1.plot(df["Frame"], df["MaxHeight"], '.', color=colors[0], label="Chamber 1 Data")
#     ax4.plot(df["Frame"], df["MaxHeight"], '.', color=colors[0], label="Cam "+ str(df["cam"].iloc[0]) + "Data")

    ax1.legend(loc="lower right")
    
    df = cam2
    coefs = poly.polyfit(df["Frame"], df["MaxHeight"], polyorder)
    x_new = np.linspace(df["Frame"].iloc[1], df["Frame"].iloc[-1], num=len(df["Frame"])*10)
    ffit = poly.polyval(x_new, coefs)
    fderiv = poly.polyval(x_new, np.polyder(coefs))

    
    ax2.plot(x_new, ffit, label="Chamber 2 (Control aka No Slat)", color=colors[1])
    ax4.plot(x_new, fderiv, label="Chamber 2 fit deriv", color=colors[1])
    #ax4.plot(x_new, ffit, label="Cam "+ str(df["cam"].iloc[0]), color=colors[1])
    ax2.plot(df["Frame"], df["MaxHeight"], '.', color=colors[1], label="Chamber 2 Data")
#     ax4.plot(df["Frame"], df["MaxHeight"], '.', color=colors[1], label="Cam "+ str(df["cam"].iloc[0]) + "Data")
    ax2.legend(loc="lower right")
    
    df = cam3
    coefs = poly.polyfit(df["Frame"], df["MaxHeight"], polyorder)
    x_new = np.linspace(df["Frame"].iloc[1], df["Frame"].iloc[-1], num=len(df["Frame"])*10)
    ffit = poly.polyval(x_new, coefs)
    fderiv = poly.polyval(x_new, np.polyder(coefs))

    ax3.plot(x_new, ffit, label="Chamber 3 (Ringed Slat)", color=colors[2])
#     ax4.plot(x_new, ffit, label="Cam "+ str(df["cam"].iloc[0]), color=colors[2])
    ax4.plot(x_new, fderiv, label="Chamber 3 fit deriv", color=colors[2])
    ax3.plot(df["Frame"], df["MaxHeight"], '.', color=colors[2], label="Chamber 3 Data")
#     ax4.plot(df["Frame"], df["MaxHeight"], '.', color=colors[2], label="Cam "+ str(df["cam"].iloc[0]) + "Data")

    ax3.legend(loc="lower right")
    
    ax4.legend(loc="lower right")
    ax1.set_xlabel("Frame (aka Time)")
    ax1.set_ylabel("Maximum Height")
    ax2.set_xlabel("Frame (aka Time)")
    ax2.set_ylabel("Maximum Height")
    ax3.set_xlabel("Frame (aka Time)")
    ax3.set_ylabel("Maximum Height")
    ax4.set_xlabel("Frame (aka Time)")
    ax4.set_ylabel("Maximum Height")
    if parab+1 < 10:
        num0s = "0"
    else:
        num0s = ""
    
    if parab+1 < 10:
        num0s = "0"
    else:
        num0s = ""
    
    plt.savefig("./poly15_wderivs-water/Parab"+num0s +"%i"%(parab+1)+".jpg", dpi=100)
```
