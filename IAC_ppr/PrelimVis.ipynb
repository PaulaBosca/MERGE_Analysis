```python
#Add a column to the concatenated frames with the maximum slice height, and the medium height
df_c1["Max"] = df_c1.filter(regex=("Slice.*_Max"),axis=1).max(axis=1)
df_c2["Max"] = df_c2.filter(regex=("Slice.*_Max"),axis=1).max(axis=1)
df_c3["Max"] = df_c3.filter(regex=("Slice.*_Max"),axis=1).max(axis=1)

df_c1["Median"] = df_c1.filter(regex=("Slice.*"),axis=1).median(axis=1)
df_c2["Median"] = df_c2.filter(regex=("Slice.*"),axis=1).median(axis=1)
df_c3["Median"] = df_c3.filter(regex=("Slice.*"),axis=1).median(axis=1)
```

```python
#plot the maximum hight over time
plt.clf()
plt.figure(dpi=200)
# df_c1.iloc[0:5000].plot(x="Time", y="Max", use_index=False, label="Cam1", kind='line')
plt.plot(df_c1["Time"], df_c1["Max"], '.', label='Cam1', alpha=0.5)
# plt.plot(df_c2["Time"], df_c2["Max"], '.', label='Cam2', alpha=0.5)
# plt.plot(df_c3["Time"], df_c3["Max"], '.', label='Cam3', alpha=0.5)
plt.xlabel("Time")
plt.ylabel("Max height")
plt.legend()
```

```python
#get the index of where Time==0 as that splits the concatenated frames into the diff parabolas 
df_c1_0s = df_c1[df_c1['Time'] == 0].index.tolist() #cam1 locations of time ==0 in dF_c1  (cam1 df)
df_c2_0s = df_c2[df_c2['Time'] == 0].index.tolist() #cam1 locations of time ==0 in dF_c1  (cam1 df)
df_c3_0s = df_c3[df_c3['Time'] == 0].index.tolist() #cam1 locations of time ==0 in dF_c1  (cam1 df)

# fig, ax = plt.subplots()  # Here we create the axes that all data will plot onto
fig,ax = plt.subplots(figsize=(12,10))
# for i in range(0,2):
# for i in range(0,len(df_c1_0s)-1):
for i in range(0,15):
    #loops through and plots all points between pairs of Time==0 (so it plots one parab at a time)
    df_c1.iloc[df_c1_0s[i]:df_c1_0s[i+1]].plot(x="Time", y="Max", 
                                               use_index=False, label="Cam 1 parab"+str(i+1), ax=ax, marker='.', color='black')
    df_c1.iloc[df_c1_0s[i]:df_c1_0s[i+1]].plot(x="Time", y="Median", 
                                               use_index=False, label="Cam 1 parab"+str(i+1), ax=ax, marker='.', color='red')
    
    df_c2.iloc[df_c2_0s[i]:df_c2_0s[i+1]].plot(x="Time", y="Max", 
                                               use_index=False, label="Cam 2 parab"+str(i+1), ax=ax, marker='.', color='blue')
    df_c2.iloc[df_c2_0s[i]:df_c2_0s[i+1]].plot(x="Time", y="Median", 
                                               use_index=False, label="Cam 2 parab"+str(i+1), ax=ax, marker='.', color='green')

    df_c3.iloc[df_c3_0s[i]:df_c3_0s[i+1]].plot(x="Time", y="Max", 
                                           use_index=False, label="Cam 3 parab"+str(i+1), ax=ax, marker='.', color='orange')
    df_c3.iloc[df_c3_0s[i]:df_c3_0s[i+1]].plot(x="Time", y="Median", 
                                           use_index=False, label="Cam 3 parab"+str(i+1), ax=ax, marker='.', color='purple')

```
