```python
## adds the extra columns indicating how many slices above the cutoff
h = 16.192    #chamber height

df_list = [df_c1p1, df_c1p2, df_c1p3, df_c1p4, df_c1p5, df_c1p6, df_c1p7, df_c1p8, df_c1p9, df_c1p10,
         df_c1p11, df_c1p12, df_c1p13, df_c1p14, df_c1p15,
           df_c2p1, df_c2p2, df_c2p3, df_c2p4, df_c2p5, df_c2p6, df_c2p7, df_c2p8, df_c2p9, df_c2p10,
          df_c2p11, df_c2p12, df_c2p13, df_c2p14, df_c2p15,
           df_c3p1, df_c3p2, df_c3p3, df_c3p4, df_c3p5, df_c3p6, df_c3p7, df_c3p8, df_c3p9, df_c3p10,
          df_c3p11, df_c3p12, df_c3p13, df_c3p14, df_c3p15]

def addingNab(df):
    df_masked = df.filter(regex='Slice[^78]*_Max')
    df["Nab25"] = df_masked.gt(0.25*h).sum(axis=1)
    df["Nab50"] = df_masked.gt(0.50*h).sum(axis=1)
    df["Nab75"] = df_masked.gt(0.75*h).sum(axis=1)
    df["Nab95"] = df_masked.gt(0.95*h).sum(axis=1)
    return df

for df in df_list: 
    df = addingNab(df)
```

```python
#creating summary df
cam_list = []
for i in range(1,4):
    cam_list = np.append(cam_list, np.full((1,15),[i]))

i=1
parab_list = []
while i <= 3:
    for j in range(1,16):
        parab_list = np.append(parab_list, j)
    i+=1

df_sum = pd.DataFrame({'cam': cam_list, 'parab': parab_list})


#getting the time it takes from the first instance to get to the percent
def timeto(df,  minslice):
    idxTime25 = (df.Nab25 >= minslice).idxmax()
    Timeto25 = df.Time.iloc[idxTime25] - df.Time.iloc[0]
    
    idxTime50 = (df.Nab50 >= minslice).idxmax()
    Timeto50 = df.Time.iloc[idxTime50] - df.Time.iloc[0]

    idxTime75 = (df.Nab75 >= minslice).idxmax()
    Timeto75 = df.Time.iloc[idxTime75] - df.Time.iloc[0]

    idxTime95 = (df.Nab95 >= minslice).idxmax()
    Timeto95 = df.Time.iloc[idxTime95] - df.Time.iloc[0]
    arr = [Timeto25, Timeto50, Timeto75, Timeto95]
    return np.array(arr)


arrofTimesto = []
for df in df_list: 
    arrofTimesto.append(timeto(df, 3))
    
arrofTimesto =  pd.DataFrame(np.array(arrofTimesto))
df_sum = pd.concat([df_sum, arrofTimesto], axis=1)
df_sum.rename(columns={0: 'Timeto25', 1: 'Timeto50', 2: 'Timeto75', 3: 'Timeto95'}, inplace=True)
```

```python
###amount of time spent above
# df_sum.drop(['frac25', 'frac50', 'frac75', 'frac95'], inplace=True, axis=1)
def timespent(df,  minslice):
    timediffframes = df.Time.iloc[1] - df.Time.iloc[0]
    tottime = df.Time.iloc[-1] - df.Time.iloc[0]

    frames25 = (df.Nab25 >= minslice).sum() -1
    timespent25 = frames25 * timediffframes
    frac25 = timespent25 / tottime
    
    frames50 = (df.Nab50 >= minslice).sum() -1
    timespent50 = frames50 * timediffframes
    frac50 = timespent50 / tottime
    
    frames75 = (df.Nab75 >= minslice).sum() -1
    timespent75 = frames75 * timediffframes
    frac75 = timespent75 / tottime
    
    frames95 = (df.Nab95 >= minslice).sum() -1
    timespent95 = frames95 * timediffframes
    frac95 = timespent95 / tottime

    arr = [frac25, frac50, frac75, frac95]
    return np.array(arr)


arrofTimespent = []
for df in df_list: 
    arrofTimespent.append(timespent(df, 2))
    
arrofTimespent =  pd.DataFrame(np.array(arrofTimespent))
df_sum = pd.concat([df_sum, arrofTimespent], axis=1)
df_sum.rename(columns={0: 'frac25', 1: 'frac50', 2: 'frac75', 3: 'frac95'}, inplace=True)

df_sum
```
